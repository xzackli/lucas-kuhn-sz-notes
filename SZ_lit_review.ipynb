## Summary of review

### The SZ Effect

Conventionally, if a photon scatters off an electron at rest, then due to recoil, a small fraction of the photon energy is transferred to the electron; this is known as the **Compton Effect** (assuming energy loss is appreciable, i.e. $E_\gamma<<m_\mathrm{e}c^2$ is not the case as is in Thompson scattering). Conversely, if an electron has a much larger energy than the photon, part of the electron energy can be transferred to the photon, which results in a mean energy gain of:
$$\Delta E_\gamma = (4k_\mathrm{B}T-E_\gamma)\frac{E_\gamma}{m_\mathrm{e}c^2}$$
where $E_\gamma$ is the incoming energy of a photon and $T$ is the themral temperature of the plasma it passes through. This is called the **Inverse Comptom Effect**.



On large scales, this inverse Compoton scattering can be seen in the Cosmic Microwave Background radiation (CMB) as electrons in the hot gas of the intracluster medium (ICM) can scatter these CMB photons. If we observe a photon that has passed through a cluster of galaxies, inverse compton scattering will change its direction i.e. scatter (though due to the fact that the CMB is isotropic, this does not affect the total number of photons we observe) and its energy in a way that it will on average have a higher frequency after scattering. As a consequence, this scattering leads to a reduced number of photons at lower energies, relative to the Planck spectrum, which are instead shifted to higher energies. However, there is one specific photon energy where the intensity of observed radiation remains unchanged corresponding to a frequency of 218 GHz. Below this frequency, the intensity is lower than that of the CMB, and for higher frequencies, the intensity is increased. 



This effect is called the **Sunyaevâ€“Zeldovich effect (SZ-effect)**. The observation of this effect allows us to pose the following idea: The CMB spectrum, measured in the direction of a galaxy cluster, deviates from a Planck spectrum; *the degree of this deviation depends on the temperature of the cluster gas and on its density, and is independent of the cluster redshift*. In other words, the SZ effect is essentially a secondary anisotropy on the CMB resulting from the interactions of CMB photons with intervening matter along the line-of-sight as they pass to detectors from the surface of last scattering.



### Components of the SZ Effect

Although the physical origin of all these SZ signals is due to simple electron scattering, each of them has slightly different spectral and spatial dependence across the cluster. In the above, I detail what is known as the **thermal SZ effect** (tSZ), i.e. when photons from the CMB scatter off electrons in the gas between galaxies, they get bumped to higher energies. In other words, the inverse Compton scattering moves them from the low-frequency region of the blackbody spectrum towards higher energies.



However, if these galaxies are in motion with respect to the expanding universe, the CMB photons get a second energy shift, either up or down depending on the relative motion of the cluster. This results in another type of SZ effect caused by called the **kinetic SZ effect** (kSZ). This effect due to the potential for the clusters to have some peculiar velocity which brings about a subsequent bulk velocity of electrons in the ICM with respect to the CMB rest frame. Suppose the peculiar velocity of a cluster is directed towards us, then the photons scattering in its intracluster gas and reaching us will be scattered by electrons which have an average velocity towards us. In other words, moving electrons can transfer some of their kinetic energy to the CMB photon field through the Doppler effect. These scattered photons thus experience on average a blueshift, which is visible in the CMB spectrum in the direction of this cluster. The kSZ effect has a different spectral signature than the tSZ effect and can thus in principle be distinguished from it. A robust measurement of the kSZ would allow a direct measurement of the line-of-sight component of the peculiar velocities of clusters. However, the expected amplitude of the kinetic SZ-effect is smaller than that of the tSZ effect by a factor of $\sim$ 10 and thus much harder to detect.



Depending on the characteristic speed of the electrons, relativistic corrections can become important, yielding the **relativistic SZ effect** (rSZ). In the case where electrons in a thermal gas have temperatures of about $kT\simeq 5$ keV, their speeds exceed that required of the non-relativistic approximation, and thus special-relativistic corrections become relevant which leads to changes of the tSZ and kSZ spectral shapes. One of the more important of these changes is that the average energy shift and broadening per scattering both increase more quickly with temperature than in the non-relativistic limit. Specifically, the overall relativistic temperature corrections cause a broadening of the tSZ and kSZ signal, with a systematic shift towards higher frequencies. These corrections (i.e. those kinematic due to the clusterâ€™s and observerâ€™s motions as well as those due to temperature) can be computed using `SZpack`.



### Observing the SZ Effect

In the Rayleighâ€“Jeans domain of the CMB spectrum (i.e. $h\nu<<k_\mathrm{B}T$), the intensity of the CMB is decreased by the SZ-effect. This change in frequecy specific intensity is equivalent to:
$$\frac{\Delta I_\nu}{I_\nu}=-2y,\qquad y = \int dl\frac{k_\mathrm{B}T}{m_\mathrm{e}c^2}\sigma_T n_\mathrm{e}$$
where $y$ is the **Compton-y parameter**, $\sigma_T$ is the Thompson cross-section, and $T,n_\mathrm{e}$ are the temeperature and number density of electrons in the ICM respectively, and $P_\mathrm{e} = n_\mathrm{e}k_\mathrm{b}T$ is the pressure due to the electrons.

The kSZ effect produces a shift in CMB intensity as follows:
$$\Delta I\approx-I_0\frac{x^4\mathrm{e}^x}{(\mathrm{e}^x-1)^2}y_\mathrm{kSZ}$$
where $I_0$ is the CMB intensity given as $I_0= 270.33\left[\frac{T_\mathrm{CMB}}{2.7255\mathrm{ K}}\right]^3\mathrm{MJy/sr}$, $x=h\nu/k_\mathrm{B}T_\mathrm{CMB}$ is a dimensionless frequency and $y_\mathrm{kSZ}$ is a kSZ analogue to the Compton-y parameter. This signal is indistinguishable from that of the hot and cold spots in the primary CMB, unless scale-dependent information or correlations with other astronomical data are exploited.

The amplitude of the spectral distortion caused by the SZ effect is given by $y$, and SZ maps essentially provide a map of $y$ as a function of angular position in the cluster. Thus, when integrating $y$ across a cluster, we can find an expression for the total SZ signal:
$$\int\mathrm{d}^2\theta y=\frac{1}{D_\mathrm{A}^2}\int\mathrm{d}^2Ry\propto\frac{1}{D_\mathrm{A}^2}\int\mathrm{d}Vn_\mathrm{e}T$$

It is important to note that since the SZ-signal $\Delta I_\nu/I_\nu$ is independent of cluster redshift, it allows us to investigate clusters at large redshifts (provided the SZ-signal is spatially resolved). This is because of the tSZ effect's near redshift independence. The spectral shape remains unchanged and the tSZ does not suffer from redshift-dimming. As a result, we can use observations of the SZ effect to identify and map massive, high-redshift galaxy clusters. Additionally, we can use observations of the SZ effect to  analyze the gas in clusters. In particular, we look to obtain more information about the spatial density and temperature distribution. of such gas.
